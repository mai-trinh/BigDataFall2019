```python
# DATA FRAME 

from pytrends.request import TrendReq
import matplotlib.pyplot as plt
import pandas as pd 
import geopandas as gpd
from descartes import PolygonPatch


import plotly.plotly as py
import plotly.graph_objs as go

# set the filepath and load in a shapefile
fp = "/Users/macboook/desktop/answer_BigData/countries_shp/countries.shp"
map_df = gpd.read_file(fp)
# check data type so we can see that this is not a normal dataframe, but a GEOdataframe
# print (map_df.head())
# set the index of the data set to be name of countries 
map_df.set_index('NAME',inplace = True)
# map_df.head()

# Log in to Google  
pytrends = TrendReq(hl='en-US', tz=360)

# The list of keywords that we want to search through 
kw_list = ['buy flight ticket']

# Build payload
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='', gprop='')

interest_by_region_df = pytrends.interest_by_region()

interest_by_region_df = interest_by_region_df.cumsum()

data_for_map = interest_by_region_df
# print (data_for_map.loc['Bahamas']['flight ticket'])
# print (data_for_map.index.tolist())

# join the geodataframe with the cleaned up dataframe
merged = map_df.join(data_for_map) # success !!!
# print (merged)

# set a variable that will call whatever column we want to visualise on the map
variable = 'buy flight ticket'

# set the range for the choropleth
vmin, vmax = 0, data_for_map.values.max()

# create figure and axes for Matplotlib
fig, ax = plt.subplots(1, figsize=(8, 4))  

# create map
merged.plot(column=variable, cmap='Spectral', linewidth=0.2, ax=ax, edgecolor='0.8')

# remove the axis
ax.axis('off')

# add a title
ax.set_title('Key word: "buy flight ticket" present - 5 years ago ', fontdict={'fontsize': '15', 'fontweight' : '3'})

# create an annotation for the data source
ax.annotate('Source: GoogleTrends' ,xy=(0.1, .08),  xycoords='figure fraction', horizontalalignment='left', verticalalignment='top', fontsize=12, color='#555555')

# Create colorbar as a legend
sm = plt.cm.ScalarMappable(cmap='Spectral', norm=plt.Normalize(vmin=vmin, vmax=vmax))

# empty array for the data range
sm._A = []

# add the colorbar to the figure
cbar = fig.colorbar(sm)

# export image 
fig.savefig("map_export_buy_flight_ticket.png", dpi=300)


```
